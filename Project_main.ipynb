# Project Description

## 1. Overview
Teaté is a digital platform with a B2B model that connects suppliers with small neighborhood grocery stores, which allows store owners the acquisition of products at convenient prices to supply their businesses. Store owners receive their supplies 48 hours after they order y  and own a digital wallet. Teaté believes in sustainable economic development, supporting small businesses and creating more collaborative and digital communities.

## 2. Problem Definition

Store owners have limited resources to supply their stores, and use their intuition to figure out which products h te q they should buy. This lack of information and  I  ppl small budget has an impact on their revenues and business, which affects the a al. Tl^’ possibility of offering a wider portfolio and satisfying demand.   

It is important to keep in mind that sales in the traditional channel represent 50% of total sales in mass consumer products, and lowering prices, will help them sell their products at competitive prices.

In order to maximize revenue and optimize resources for acquiring merchandise, the idea is to find optimal combinations between products regarding quantity and price to guarantee enough stock and avoid oversupply in the stores.

## 3. Datasets to use

The data sets for the project will be:

- Historical transactions of the app Teaté between stores and suppliers from January to July 2020. These transactions contain information about each product sold by the app.The main variables of the data set are related to the store (ID, location), the products (ID, quantity, cost) and also the dates of orders. 
- Public socio-economic data provided by DANE.



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

months = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio']
list_of_df = []
for m in months:
    list_of_df.append(pd.read_csv('data/Teaté - Venta '+m+' 2020.csv', encoding='ISO-8859-1', sep=';', low_memory=False))
```

```python
# Now lets concatenate our list of DFs
final_df = pd.concat(list_of_df)

final_df.head()
```

```python
final_df.to_csv('DF_completo.csv', index=False)
```
